```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv('Globalweather.csv')
```

```python
df
```

```python
for i in range(0,df.shape[0]-6):
    df.loc[df.index[i+6],'SMA_7'] = np.round(((df.iloc[i,1]+ df.iloc[i+1,1] +df.iloc[i+2,1] + df.iloc[i+3,1] + df.iloc[i+4,1] + df.iloc[i+5,1] + df.iloc[i+6,1])/7),1)
```

```python
df
```

```python
dfny = pd.read_csv('NYweather.csv')
dfny
```

```python
for i in range(0,dfny.shape[0]-6):
    dfny.loc[dfny.index[i+6],'SMA_7'] = np.round(((dfny.iloc[i,3]+ dfny.iloc[i+1,3] +dfny.iloc[i+2,3] + dfny.iloc[i+3,3] + dfny.iloc[i+4,3] + dfny.iloc[i+5,3] + dfny.iloc[i+6,3])/7),3)
```
